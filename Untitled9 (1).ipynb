```python
def genderiser(x):
    if x==0:
        return 'female'
    return 'male'
dataset2['Gender']=dataset2['Gender'].apply(genderiser)

def very(x):
    if x==0.0:
        return 'Not Set'
    elif x==1.0:
        return 'Income unverified'
    elif x==2.0:
        return 'Income unverified cross-referenced by phone'
    elif x==3.0:
        return 'Income verified'
    else:
        return 'Income and expenses verified'
dataset2['VerificationType']=dataset2['VerificationType'].apply(very)

def lim(x):
    if x==1:
        return 'estonian'
    elif x==2:
        return 'English'
    elif x==3:
        return "Russian"
    elif x==4:
        return "Finnish"
    elif x==5:
        return "German"
    elif x==6:
        return "Spanish"
    elif x==9:
        return "Slovakian"
    else:
        return "Others"
 
dataset2['LanguageCode']=dataset2['LanguageCode'].apply(lim)

def edu(x):
    if x==-1:
        return 'Not Present'
    elif x==0:
        return 'Not Present'
    elif x==1:
        return 'Primary education'
    elif x==2:
        return 'Basic education'
    elif x==3:
        return 'Vocational education'
    elif x==4:
        return 'Secondary education'
    elif x==5:
        return 'Higher education'

    
dataset2['Education']=dataset2['Education'].apply(edu)

def loan(x):
    if x==-1:
        return 'Loan consolidation'
    if x==0:
        return 'Real estate'
    if x==1:
        return 'Home improvement'
    if x==2:
        return 'Business'
    if x==3:
        return 'Business'
    if x==4:
        return 'Education'
    if x==5:
        return 'Travel'
    if x==6:
        return 'Vehicle'
    if x==7:
        return 'Other'
    if x==8:
        return 'Health'
    if x==101:
        return 'Working capital financing'
    if x==102:
        return 'Purchase of machinery equipment '
    if x==104:
        return 'Accounts receivable financing'
    if x==106:
        return 'Acquisition of stocks'
    if x==107:
        return 'Acquisition of stocks'
    if x==108:
        return 'Acquisition of real estate'
    if x==110:
        return 'Other business All codes in format 1XX are for business loans that are not supported since October 2012'
    
dataset2['UseOfLoan']=dataset2['UseOfLoan'].apply(loan)        

#I will finish up on the rest of the category variables
```
